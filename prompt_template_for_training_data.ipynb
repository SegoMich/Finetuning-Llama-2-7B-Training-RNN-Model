```python
import pandas as pd
#read chat csv to df
main_df = pd.read_csv("chat.csv")
main_df.sample(5)
```

```python
#chat prompt templates
promptlst = []
for i in range(len(main_df)):
    prompt = "<s><INST>" + str(main_df['input'][i]) + "</INST>" + str(main_df['response'][i]) + "</s>"
    promptlst.append(prompt)
```

```python
# Questions related to digital transformation with additional details
questions = [
    "What are the key drivers for digital transformation at ABC Consultancy Company in Cityville, XYZ?",
    "How can we assess ABC Consultancy Company's digital readiness given its location at 123 Main Street, Cityville, XYZ?",
    "What are the main challenges ABC Consultancy Company may face in implementing digital transformation?",
    "What are the benefits for ABC Consultancy Company in Cityville, XYZ, of embracing digital transformation?",
    "How can ABC Consultancy Company develop a digital transformation strategy tailored to its operations at 123 Main Street, Cityville, XYZ?",
    "What technologies are essential for ABC Consultancy Company's digital transformation in Cityville, XYZ?",
    "How can ABC Consultancy Company ensure cybersecurity in its digital transformation efforts?",
    "What role does leadership play at ABC Consultancy Company in driving digital transformation?",
    "How can ABC Consultancy Company measure the success of its digital transformation initiatives?",
    "What are some best practices for managing change during digital transformation at ABC Consultancy Company?"
]

# Responses corresponding to each question with additional details
responses = [
    "Key drivers for digital transformation at ABC Consultancy Company include changing customer expectations, technological advancements, competitive pressures, and regulatory requirements.",
    "ABC Consultancy Company's digital readiness can be assessed through various factors such as technology infrastructure, organizational culture, workforce skills, and digital strategy alignment.",
    "Challenges ABC Consultancy Company may face in implementing digital transformation include resistance to change, legacy systems, data privacy concerns, talent shortages, and budget constraints.",
    "Embracing digital transformation can lead to improved customer experiences, increased operational efficiency, faster time-to-market, better decision-making, and competitive advantage for ABC Consultancy Company.",
    "Developing a digital transformation strategy for ABC Consultancy Company involves assessing current capabilities, defining goals and objectives, identifying technology investments, and establishing a roadmap for implementation.",
    "Essential technologies for ABC Consultancy Company's digital transformation may include cloud computing, big data analytics, artificial intelligence, Internet of Things (IoT), and robotic process automation (RPA).",
    "Cybersecurity in digital transformation efforts at ABC Consultancy Company can be ensured through regular risk assessments, robust security protocols, employee training, encryption, and compliance with regulations.",
    "Leadership at ABC Consultancy Company plays a crucial role in driving digital transformation by setting a clear vision, fostering innovation, empowering employees, and allocating resources effectively.",
    "Success of digital transformation initiatives at ABC Consultancy Company can be measured through key performance indicators (KPIs) such as revenue growth, customer satisfaction, operational efficiency, and agility.",
    "Best practices for managing change during digital transformation at ABC Consultancy Company include effective communication, stakeholder engagement, training and upskilling programs, and establishing a change management framework."
]
```

```python
for i in range(len(questions)):
    prompt = "<s><INST>" + str(questions[i]) + "</INST>" + str(responses[i]) + "</s>"
    promptlst.append(prompt)
```
